```python
import pandas as pd
```

```python
df_posts = pd.read_csv('data/wpru_posts.csv')
df_comps = pd.read_csv('data/preprocessed_funding.csv')
df_terms = pd.read_csv('data/wpru_terms.csv')
df_term_relations = pd.read_csv('data/wpru_term_relationships.csv')
```

```python
df_related = pd.merge(df_posts,df_term_relations, left_on = 'ID', right_on = 'object_id',how = 'inner').drop('ID',axis=1)
df_complete = pd.merge(df_related,df_terms, left_on = 'term_taxonomy_id', right_on = 'term_id',how = 'inner').drop('term_taxonomy_id',axis=1)
```
